```python
import time
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
```

```python
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")


data = []

with Chrome(options=options) as driver:
    wait = WebDriverWait(driver, 15)
    driver.get("https://www.youtube.com/watch?v=epnC7Fwwd-E")

    for item in range(10):
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(10)

    comments = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".ytd-comment-thread-renderer")))

    for comment in comments:
        try:
            # Extract username
            username = comment.find_element(By.CSS_SELECTOR, "#author-text").text

            # Extract date
            date = comment.find_element(By.CSS_SELECTOR, "#published-time-text").text

            # Extract like count
            like_element = comment.find_element(By.CSS_SELECTOR, "#vote-count-middle")
            like_count = like_element.text if like_element.text else "0"  # Defaults to "0" if empty

            # Extract comment text
            comment_text = comment.find_element(By.CSS_SELECTOR, "#content-text").text
            # Append data to list
            data.append({
                "username": username,
                "date": date,
                "like_count": like_count,
                "comment": comment_text
            })
        except Exception as e:
            print(f"An error occurred: {e}")

# Convert to DataFrame
df = pd.DataFrame(data, columns=['username', 'comment', 'date', 'like_count'])
print(df)
```

```python
df.to_csv("data_komentar.csv", index=False)
print("File 'youtube_comments.csv' has been saved successfully.")
```
