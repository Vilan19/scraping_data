```python
import time
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
```

```python
#set up untuk menjalankan browser
options = Options()
options.add_argument("--headless") #scraping di latar belakang tanpa membuka jendela browser
options.add_argument("--no-sandbox") #menjalankan Chrome tanpa akses root
options.add_argument("--disable-dev-shm-usage") #mengarahkan Chrome agar menggunakan memori disk


data = []

with Chrome(options=options) as driver:
    wait = WebDriverWait(driver, 15) #waktu tunggu maksimum untuk perintah wait.until(...)
    driver.get("https://www.youtube.com/watch?v=epnC7Fwwd-E")

    for item in range(10): #jumlah loop yg dilakukan agar lebih banyak komentar dimuat
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(5)

    #menunggu hingga semua elemen komentar muncul
    comments = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".ytd-comment-thread-renderer")))

    for comment in comments:
        try:
            #menemukan elemen yang berisi nama pengguna (username)
            username = comment.find_element(By.CSS_SELECTOR, "#author-text").text

            #menemukan waktu pada komentar
            date = comment.find_element(By.CSS_SELECTOR, "#published-time-text").text

            #menemukan jumlah like pada komentar
            like_element = comment.find_element(By.CSS_SELECTOR, "#vote-count-middle")
            like_count = like_element.text if like_element.text else "0"  # Defaults to "0" if empty

            #menemukan teks komentar
            comment_text = comment.find_element(By.CSS_SELECTOR, "#content-text").text

            # Append data to list
            data.append({
                "username": username,
                "date": date,
                "like_count": like_count,
                "comment": comment_text
            })
        except Exception as e:
            print(f"An error occurred: {e}")

# Convert to DataFrame
df = pd.DataFrame(data, columns=['username', 'comment', 'date', 'like_count'])
print(df)
```

```python
df.to_csv("data_komentar.csv", index=False)
print("File 'youtube_comments.csv' has been saved successfully.")
```
